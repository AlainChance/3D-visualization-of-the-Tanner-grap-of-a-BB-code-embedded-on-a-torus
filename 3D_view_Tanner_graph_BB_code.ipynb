```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load and resize the image
img = Image.open('tanner_graph.png')  # Tanner graph image taken from Andrew & Gambetta, Jay & Maslov, Dmitri & Rall, Patrick & Yoder, Theodore. (2024),
                                      # High-threshold and low-overhead fault-tolerant quantum memory. Nature. 627. 778-782. 10.1038/s41586-024-07107-7
img = img.resize((600, 600))          # Resample if needed
img = np.array(img) / 255.0           # Normalize pixel values

# Parameters for torus
R, r = 2, 1   # Major and minor radius
u = np.linspace(0, 2 * np.pi, img.shape[1])
v = np.linspace(0, 2 * np.pi, img.shape[0])
u, v = np.meshgrid(u, v)

# Map texture coordinates to torus
X = (R + r * np.cos(v)) * np.cos(u)
Y = (R + r * np.cos(v)) * np.sin(u)
Z = r * np.sin(v)

# Plot the image texture on the torus
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=img, shade=False)

ax.set_axis_off()
ax.set_title('3D visualization of the Tanner graph of a BB code embedded on a torus - By Alain Chancé')
plt.tight_layout()
plt.show()

print("2D Tanner graph representation for the Gross code mapped on a 3D Torus")
print("Authored by Alain Chancé on June 24, 2025.\n")

print("Credit: Source image taken from the following article which is licensed under a Creative Commons Attribution 4.0 International License:")
print("Bravyi, Sergey & Cross, Andrew & Gambetta, Jay & Maslov, Dmitri & Rall, Patrick & Yoder, Theodore.")
print("(2024). High-threshold and low-overhead fault-tolerant quantum memory. Nature. 627. 778-782. 10.1038/s41586-024-07107-7.")
print("Fig. 1 | b, Tanner graph of a BB code with parameters [[14 4, 12, 12]] embedded into a torus.")
print("https://www.researchgate.net/figure/Tanner-graphs-of-surface-and-BB-codes-a-Tanner-graph-of-a-surface-code-for-comparison_fig3_379336526\n")

print("Mapping achieved with Python code generated by Microsoft Copilot, with prompt: '3D visualization of a given image mapped onto a torus'")
print("https://copilot.microsoft.com/shares/PRmF8VPhYCX1wHBH25eJz")
```
